```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
plt.rcParams.update({'font.size': 12})
plt.figure(figsize=[6,4.5])

for i in range(100):
    err_logPS = np.load(f'outputs_AISTATS/Cardiac_err_log_PSCT_{i}.npy',allow_pickle=True)
    err_logRS = np.load(f'outputs_AISTATS/Cardiac_err_log_RSCT_{i}.npy',allow_pickle=True)
    nf_PS = np.load(f'outputs_AISTATS/Cardiac_nf_PSCT_{i}.npy',allow_pickle=True)
    nf_RS = np.load(f'outputs_AISTATS/Cardiac_nf_RSCT_{i}.npy',allow_pickle=True)
       
    plt.plot(nf_RS[np.where((nf_RS>=100)&(nf_RS<500))], 
             err_logRS[np.where((nf_RS>=100)&(nf_RS<500))],color='C0',alpha=0.2,lw=1)
    plt.plot(nf_PS[np.where((nf_PS>=100)&(nf_PS<500))], 
             err_logPS[np.where((nf_PS>=100)&(nf_PS<500))],color='C1',alpha=0.2,lw=1)

s1 = plt.scatter(592, 0.22, edgecolor='C2', facecolor='None', s=15, label='RP-HOSVD')
s2 = plt.scatter(592, 0.0085, edgecolor='C3', facecolor='None', s=15, label='RP-STHOSVD') 
s3 = plt.scatter(592, 2.4, edgecolor='C4', facecolor='None', s=15, label='R-PET')
s4 = plt.scatter(592, 0.30, edgecolor='C5', facecolor='None', s=15, label='R-ST')
s5 = plt.scatter(592, 0.14, edgecolor='C6', facecolor='None', s=15, label='R-HOID')
s6 = plt.scatter(592, 0.20, edgecolor='C7', facecolor='None', s=15, label='RP-HOOI')
s7 = plt.scatter(592, 0.38, edgecolor='C8', facecolor='None', s=15, label='R-SCTT')

RS_lines = mlines.Line2D([], [], color='C0',alpha=0.3,lw=1,label='R-SCT trials')
PS_lines = mlines.Line2D([], [], color='C1',alpha=0.3,lw=1,label='P-SCT trials')

plt.legend(handles=[s1,s2,s3,s4,s5,s6,s7,RS_lines,PS_lines],ncol=2,fontsize=9,loc=9)

plt.ylabel('log err')
plt.xlabel('# of total row vectors')
plt.title('Cardiac MRI learning curve')
plt.xticks([30,100,150,200,250,300,350,400,450,500,592],[30,100,150,200,250,300,350,400,450,500,'592\n(all)'])
plt.yscale('log')
plt.xlim([100,600])
plt.ylim([0.005,100])

plt.tight_layout()
#plt.savefig('../plots/supp_cardiac_lc1_AISTATS.png',dpi=600)
plt.show()
```
