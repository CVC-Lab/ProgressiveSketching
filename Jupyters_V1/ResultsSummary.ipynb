```python
import numpy as np
from numpy.linalg import svd, norm
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import scipy.io
from scipy import interpolate
```

```python
plt.figure(figsize=[10,3])

err_logPS_interp = np.zeros((10,39))
err_logRS_interp = np.zeros((10,39))
for i in range(10):
    err_logPS = np.load(f'../hpc_codes/outputs/Cardiac_err_logPS_{i}.npy',allow_pickle=True)
    err_logRS = np.load(f'../hpc_codes/outputs/Cardiac_err_logRS_{i}.npy',allow_pickle=True)
    nf_PS = np.load(f'../hpc_codes/outputs/Cardiac_nf_PS_{i}.npy',allow_pickle=True)
    nf_RS = np.load(f'../hpc_codes/outputs/Cardiac_nf_RS_{i}.npy',allow_pickle=True)
    
    f_PS = interpolate.interp1d(nf_PS, err_logPS)
    err_logPS_interp[i] = f_PS(np.linspace(120,500,39))
    f_RS = interpolate.interp1d(nf_RS, err_logRS)
    err_logRS_interp[i] = f_RS(np.linspace(120,500,39))
    
    plt.plot(nf_PS, np.log(err_logPS),color='C0',alpha=0.3,lw=1)
    plt.plot(nf_RS, np.log(err_logRS),color='C1',alpha=0.3,lw=1)

err_logPS_interp = np.sort(err_logPS_interp,axis=0)
err_logRS_interp = np.sort(err_logRS_interp,axis=0)

plt.plot(np.linspace(120,500,39), np.log(np.mean(err_logPS_interp[1:-1,:],axis=0)),color='C0',lw=2)
plt.plot(np.linspace(120,500,39), np.log(np.mean(err_logRS_interp[1:-1,:],axis=0)),color='C1',lw=2)

plt.vlines([200,300,400,500],plt.gca().get_ylim()[0],plt.gca().get_ylim()[1],alpha=0.3,linestyle='--',lw=1)

PS_lines = mlines.Line2D([], [], color='C0',alpha=0.3,lw=1,label='PS trials')
RS_lines = mlines.Line2D([], [], color='C1',alpha=0.3,lw=1,label='RS trials')
PS_mean = mlines.Line2D([], [], color='C0', lw=2, label='PS_mean')
RS_mean = mlines.Line2D([], [], color='C1', lw=2, label='RS_mean')
plt.legend(handles=[PS_lines,RS_lines,PS_mean,RS_mean])

plt.ylabel('log err')
plt.xlabel('# of samples')
plt.title('Cardiac dataset SketchyCoreTucker learning curve')
plt.xticks([120,150,200,250,300,350,400,450,500])
plt.xlim([110,510])
plt.ylim([-10,15])
plt.tight_layout()
plt.savefig('../plots/supp_cardiac_lc.png',dpi=300)
plt.show()
```
